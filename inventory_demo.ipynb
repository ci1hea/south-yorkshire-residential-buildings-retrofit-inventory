```python
import geopandas as gpd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sysc_buildings = gpd.read_file('sysc_buildings_retrofit_inventory.gpkg')
```

```python
sysc_buildings.describe()
```

```python
sysc_buildings.groupby('Archetype', as_index=False).mean()
```

# Average Annual Carbon Savings by Archetype

```python
sns.barplot(data = sysc_buildings.groupby('Archetype', as_index=False).mean(),
            x='Archetype', y='Annual Carbon Savings, System Retrofit [tCO2e]')
plt.ylim(0,7)

plt.figure()
sns.barplot(data = sysc_buildings.groupby('Archetype', as_index=False).mean(),
            x='Archetype', y='Annual Carbon Savings, Fabric Retrofit [tCO2e]')
plt.ylim(0,7)

plt.figure()
sns.barplot(data = sysc_buildings.groupby('Archetype', as_index=False).mean(),
            x='Archetype', y='Annual Carbon Savings, Whole-house Retrofit [tCO2e]')
plt.ylim(0,7)
```

# Average Embodied Carbon Payback by Archetype

```python
sns.barplot(data = sysc_buildings.groupby('Archetype', as_index=False).mean(),
            x='Archetype', y='Embodied Carbon Payback, System Retrofit [years]')
plt.ylim(0,3)

plt.figure()
sns.barplot(data = sysc_buildings.groupby('Archetype', as_index=False).mean(),
            x='Archetype', y='Embodied Carbon Payback, Fabric Retrofit [years]')
plt.ylim(0,3)

plt.figure()
sns.barplot(data = sysc_buildings.groupby('Archetype', as_index=False).mean(),
            x='Archetype', y='Embodied Carbon Payback, Whole-house Retrofit [years]')
plt.ylim(0,3)
```
